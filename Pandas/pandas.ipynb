## Making changes to our data

```python
df.head(5) # with head method you can specify how many data you wanted to print
```

```python
df["Total"] = df["HP"] + df["Attack"]+df["Defense"]+df["Sp. Atk"]+df["Sp. Def"]+df["Speed"] # you can create a complete different column , here we create Total column adding other column
```

```python
df.head(5)
```

```python
df = df.drop(columns=["Total"]) # you can drop the total column
df.head(5)
```

```python
df["Total"] = df.iloc[:,4:10].sum(axis=1) #  you can also use iloc to get the row with quantitative data
df.head(5)
```

```python
# rearrange column
cols = list(df.columns.values) # it will give us list of all headers
df = df[cols[0:4] + [cols[-1]]+cols[4:12]] # you rearrange column just adding them up
print(df.head(5))
```

## Filtering Data

```python
df.loc[df["Type 1"] == "Grass"] # we can get specific type of data using loc method
```

```python
df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison")] # you can also multuple condition (In pandas we use "&"" not "and" )
```

```python
new_df = df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison")]
```

```python
new_df
```

```python
new_df.reset_index() # for reseting index
```

```python
#let's filter out specific types of name
df.loc[df["Name"].str.contains("Mega")] # it's gives up all the name contain mega
```

```python
# drop name that contains mega
df.loc[~df["Name"].str.contains("Mega")]
```

```python
# use regular expression for filtering data
import re # importing regex
df.loc[df["Type 1"].str.contains("fire|grass",flags=re.I,regex=True)]
```

```python
df.loc[df["Name"].str.contains("^pi[a-z]",flags=re.I,regex=True)] # all the name start with "pi"
```

```python
df.loc[df["Type 1"] == "Fire", "Type 1"] = "Flamer" # change specific qualitative data, here we change fire to flamer
```

```python
df
```

```python
# change multiplate parameter at time
df.loc[df["Total"] > 500, ["Generation","Legendary"]] = "TEST VALUE"
```

```python
df
```

```python
df.loc[df["Total"] > 500, ["Generation","Legendary"]] = ["Test 1", "Test 2"]
```

## Aggregate Statistics (Groupby)

```python
df = pd.read_csv("modified.csv")

# find avarage of special type of data

df.groupby(["Type 1"]).mean().sort_values("Defense" ,ascending=False)
```

```python
df.groupby(["Type 1"]).sum() # sum of colums data
```

```python
df.groupby(["Type 1"]).count()
```
